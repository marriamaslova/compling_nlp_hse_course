Лемматизируем при помощи mystem

```python
def mystem_lemmatization(text):
  lemmas = []
  for word in mystem.lemmatize(text):
    if word.isalpha():
      lemmas.append(word)
  return lemmas
```

Токенизируем при помощи mystem

```python
def mystem_tokenization(text):
  tokens = []
  for word in mystem.analyze(text):
    if word['text'].isalpha():
      tokens.append(word['text'])
  return tokens
```

```python
for_pymorphy = mystem_tokenization(text_for_lemmatization)
for_pymorphy
```

Лемматизируем при помощи pymorphy

```python
lemmas_pm = []
for elem in for_pymorphy:
  res = pm.parse(elem)[0].normal_form
  lemmas_pm.append(res)
lemmas_pm
```
